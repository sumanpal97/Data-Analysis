## This is just a Demo Analysis on toy datasets which will available from sklearn documentation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution Plot

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['tip'],kde=True,bins=60)
```

# JointPlot

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'])
```

# PairPlot

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='day',palette='coolwarm')
```

# KDE and Rug Plot

```python
sns.rugplot(tips['total_bill'])
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```

# Categorial Plots

Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

Count Plot

```python
sns.countplot(x='sex',data=tips)
```

Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,dodge=True,hue='sex')#dodge is same as split keyword
```

Swarm PLot


```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

Cat Plot or Factor Plot

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar') #catplot is same as factorplot
```

# GRID

Pair Grid

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

Facet Grid

```python
tips.head()
```

```python
g=sns.FacetGrid(row='time',col='smoker',data=tips)
g.map(sns.distplot,'total_bill')
```

```python
x=sns.FacetGrid(row='day',col='sex',data=tips)
x.map(plt.scatter,'total_bill','tip')
```

Regression Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill',y='tip',data=tips,col='time')
```
